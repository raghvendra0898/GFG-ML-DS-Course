```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
s = np.linspace(0,4*np.pi, 100)
sin_x = np.sin(s)
cos_x = np.cos(s)
plt.plot(s, sin_x, label = 'Sine')
plt.plot(s, cos_x, label = 'Cosine')

plt.legend(loc = 'upper right')
plt.title('Sine wave')
plt.show()
```

```python
s = np.linspace(0,4*np.pi, 100)
sin_x = np.sin(s)
cos_x = np.cos(s)
plt.plot(s, sin_x, '*')
plt.plot(s, cos_x, '-')

plt.legend(loc = 'upper right')
plt.title('Sine wave')
plt.show()
```

```python
plt.plot(s, sin_x, label='Sine')
plt.plot(s, cos_x, label = 'Cosine')

plt.title('Sine and Cosine graph')

fig = plt.figure()
fig.savefig('Sine and Cosine graph.jpeg')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.random(100)

colors = rng.randn(100)
sizes = rng.randn(100)*1000

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.7, cmap = 'viridis', edgecolors= 'black')
plt.show()
```

```python
cars = ['Audi','BMW','FORD','Tesla','Jaguar']
data = [450,385,35,1,200]
explode = [0,.2,0,.2,.5]
fig = plt.figure(figsize=(7,7))

plt.pie(data, labels=cars, explode= explode, shadow=True, autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sine wave')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cosine wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tangent wave')

y = np.random.randn(1000)
ax[1,1].hist(y, bins=40)
plt.tight_layout()
ax[1,1].set_title('Random Distribution')
```
